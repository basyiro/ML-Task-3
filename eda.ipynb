Let's see the distribution of numerical values through line graphs. We would see the survival rates of the patients at each value of a feature.

E.g., at 50 years old (__Age__ = 50), how many people survived

```python
# Check how many unique values per feature for the overall, survived and not_survive
for feature in numerical_feature_list:
    overall_len = len(patients_df[feature].value_counts().sort_index())
    survived_len = len(survived_df[feature].value_counts().sort_index())
    not_survive_len = len(not_survive_df[feature].value_counts().sort_index())
    print(
        f"{feature} has {overall_len} unique values in total, with {survived_len} unique values for survived and {not_survive_len} unique values for not survive entries"
    )
```

From this, we see that some features have no survivors at particular values (eg for __Sodium__, we have 27 different sodium values and survivors only compromise 22 of these values)

```python
# To accomodate for the difference in length, we fill the missing unique values for survive/ not_survive
# from the overall unique values.

def make_same_length(feature):
    overall = patients_df[feature].value_counts().sort_index()
    overall_copy = overall.copy(deep=True) # Copy is set to 0 and combined with survived and not_survive
    survived = survived_df[feature].value_counts().sort_index()
    not_survive = not_survive_df[feature].value_counts().sort_index()
    
    for entry in range(len(overall_copy)):
        overall_copy.iloc[entry] = 0

    survived = survived.combine(overall_copy, max, fill_value = 0)
    not_survive = not_survive.combine(overall_copy, max, fill_value = 0)
    
    return survived, not_survive, overall
```

```python
# Compare the survival rate of the feature at each value against the overall population
def survival_rate(feature):
    survived_series, not_survive_series, overall_series = make_same_length(feature)
    
    # Convert the values to percentages for easier comparison
    for count_index in range(len(overall_series)):
        survived_series.iloc[count_index] = survived_series.iloc[count_index]/overall_series.iloc[count_index]
        not_survive_series.iloc[count_index] = not_survive_series.iloc[count_index]/overall_series.iloc[count_index]
        
    return survived_series, not_survive_series
```

We now compare the different features to see how they contribute to the survival rate.
1. The first graph shows the distribution of of the survivors vs non-survivors over the total data
2. The second graph shows the survival rate across the (increasing) values of a feature.

```python
fig, axs = plt.subplots(ncols=2, nrows=5, figsize=(20, 20))
plt.subplots_adjust(right=1.1, top=5)
graph_counter = 1

for feature in numerical_feature_list:
    plt.subplot(10, 2, graph_counter)
    patients_df[feature].value_counts().sort_index().plot(kind="line")
    survived_df[feature].value_counts().sort_index().plot(kind="line")
    not_survive_df[feature].value_counts().sort_index().plot(kind="line")
    plt.legend(
        ["Overall", "Survived", "Did not survive"],
        loc="upper center",
        prop={"size": 18},
    )

    plt.xlabel("{}".format(feature), size=20, labelpad=15)
    plt.ylabel("Patient Count", size=20, labelpad=15)
    plt.tick_params(axis="x", labelsize=20)
    plt.tick_params(axis="y", labelsize=20)
    plt.title(f"Distribution for {feature}", size=20, y=1.05)
    graph_counter += 1

    plt.subplot(10, 2, graph_counter)
    survived_series, not_survive_series = survival_rate(feature)
    survived_series.plot(kind="line", label="survive")

    plt.xlabel("{}".format(feature), size=20, labelpad=15)
    plt.ylabel("Survival Rate", size=20, labelpad=15)
    plt.tick_params(axis="x", labelsize=20)
    plt.tick_params(axis="y", labelsize=20)

    plt.legend(["Survived"], loc="upper center", prop={"size": 18})
    plt.title(f"Survival rate for {feature}", size=20, y=1.05)
    graph_counter += 1

plt.show()
```

From the above, some major features with clear trends that could've contributed to survival rates are:
1. __Hemoglobin__ (lower hemoglobin tends to survive)
2. __Weight__ (higher weight tends to survive)

## 2.4 Encoding Categorical Features
We then encode categorical features using one-hot encoding for machine learning purposes. We define 3 different dataframes depending on how strict we want the columns to be from the EDA above.

```python
strict_relationship_df = patients_df.drop(["Favorite color", "Height", "Age", "Platelets", "Creatine phosphokinase"], axis = 1)
strict_relationship_df.name = "Strict Relationship"

loose_relationship_df = patients_df.drop(["Favorite color", "Height"], axis = 1)
loose_relationship_df.name = "Loose Relationship"

no_drop_df = patients_df
no_drop_df.name = "No Drop"

dataframe_list = [strict_relationship_df, loose_relationship_df, no_drop_df]
```

```python
encoded_features = []
one_hot_encoded_dataframe_list = []


def one_hot_encoding(dataframe):
    for feature in categorical_feature_list:
        try:
            encoded_feat = (
                OneHotEncoder()
                .fit_transform(dataframe[feature].values.reshape(-1, 1))
                .toarray()
            )
            n = dataframe[feature].nunique()
            cols = ["{}_{}".format(feature, n) for n in range(1, n + 1)]
            encoded_df = pd.DataFrame(encoded_feat, columns=cols)
            encoded_df.index = dataframe.index
            encoded_features.append(encoded_df)
        except:
            print(f'"{feature}" no longer in dataframe {dataframe.name}')
    dataframe = pd.concat([dataframe, *encoded_features], axis=1)
    return dataframe


strict_relationship_df = one_hot_encoding(strict_relationship_df)
loose_relationship_df = one_hot_encoding(loose_relationship_df)
no_drop_df = one_hot_encoding(no_drop_df)
```

Now, we drop the categorical features as we have one-hot encoded them, and view how the final dataframe looks like (using `no_drop_df` as an example)

# 3 Models

We define 3 models, neural network classifier, random forest classifier and adaboost classifier

```python
model_list = []

# Neural Net Classifier
mlp_clf = MLPClassifier(
    solver="adam",
    batch_size=1000,
    alpha=1e-5,
    activation="tanh",
    max_iter=500,
    early_stopping=False,
    verbose=True,
    hidden_layer_sizes=(50, 50),
    random_state=100,
)
model_list.append(mlp_clf)

# Random Forest
rf_clf = RandomForestClassifier(
    criterion="gini",
    n_estimators=1750,
    max_depth=7,
    min_samples_split=6,
    min_samples_leaf=6,
    max_features="auto",
    oob_score=True,
    random_state=13,
    n_jobs=-1,
    verbose=1,
)
model_list.append(rf_clf)

# AdaBoost Classifier
ada_clf = AdaBoostClassifier(
DecisionTreeClassifier(max_depth=50),
n_estimators=250,
learning_rate=1,
random_state=100,
)
model_list.append(ada_clf)
```

We split our dataset into training and testing using `train_test_split` then use `StandardScaler` to standardize our features for each of the dataset.

```python
# For strict and loose df
drop_columns = ["Survive", "ID", "Gender", "Smoke", "Diabetes", "Ejection Fraction"]

# For no_drop_df
drop_columns_no_drop = [
    "Survive",
    "ID",
    "Gender",
    "Smoke",
    "Diabetes",
    "Ejection Fraction",
    "Favorite color",
]


def modelling(dataframe, drop_columns):
    train, test = train_test_split(
        dataframe, test_size=0.3, random_state=42, shuffle=True
    )
    X_train = StandardScaler().fit_transform(train.drop(columns=drop_columns))
    y_train = train["Survive"].values
    X_test = StandardScaler().fit_transform(test.drop(columns=drop_columns))
    y_test = test["Survive"].values

    results = defaultdict(dict)

    for model in model_list:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        model_name = type(model).__name__
        results[model_name]["Accuracy"] = f"{accuracy_score(y_test, y_pred)*100:.2f}%"
        results[model_name]["F1 Score"] = f"{f1_score(y_test, y_pred):.2f}"

    for k, v in results.items():
        print(f"{k}")
        for k, v in v.items():
            print(f"{k} - {v}")
        print("\n")
```
